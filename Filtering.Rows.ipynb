```python
import pandas as pd
```

```python
laptops = pd.read_csv(r"D:\Data Analytics\Python\laptop_prices.csv", encoding = 'ISO-8859-1')
laptops
```

```python
laptops[laptops['Company'] == 'Apple'] #filtering where company is apple. 
```

```python
laptops[(laptops['Company'] == 'Apple') & (laptops['Inches'] == 13.3)]
```

```python
laptops[(laptops['Company'] == 'Apple') | (laptops['Inches'] == 13.3)] #the | means 'OR' as opposed to AND '&'. 
```

```python
laptops[laptops['Inches'] > 13.5] #filters for laptops with inches greater than 13.5. 
#note you can use >=,  hence greater or equal to. 
#'               ' <=, hence less than or equal to. 
#'                ' != 13.3, hence Does Not equal to 13.3. 
```

```python
#Show 5 HP laptops that are above 2000 dollars with a 15.6 inch screen. 
#Return the product and price of the laptops.

laptops[(laptops['Price_usd'] > 2000) & (laptops['Inches'] == 15.6)]
#now only return product and price. 
```

```python
#the below filters for product and price. 
laptops[(laptops['Price_usd'] > 2000) & (laptops['Inches'] == 15.6)][['Product', 'Price_usd']]
#now we need to filter for only 5 HP laptops. 
```

```python
laptops[(laptops['Price_usd'] > 2000) & (laptops['Inches'] == 15.6)][['Product', 'Price_usd']][:5]
#gives top 5. 
```
